```python
import numpy as np
import pywt
import pywt.data
import matplotlib.image as npimage
import matplotlib.pyplot as plt
```

```python
img = npimage.imread("geo.png")
```

```python
imb3 = img[:,:,1]
```

```python
coefficient = pywt.dwt2(imb3,'db3',mode='periodization')
cA, (cH, cV,cD) = coefficient
```

```python
imgr = pywt.idwt2(coefficient,'db3',mode='periodization')
imgr = np.uint8(imgr)
```

```python
plt.figure(figsize=(30,30))
plt.subplot(2,2,1)
plt.imshow(cA,cmap=plt.cm.gray)
plt.title("Approximation coefficient", fontsize=30)

plt.subplot(2,2,2)
plt.imshow(cH,cmap=plt.cm.gray)
plt.title("Horizontal Detailed coefficient", fontsize=30)

plt.subplot(2,2,3)
plt.imshow(cV,cmap=plt.cm.gray)
plt.title("Vertical Detailed coefficient", fontsize=30)

plt.subplot(2,2,4)
plt.imshow(cD,cmap=plt.cm.gray)
plt.title("Digonal Detailed coefficient", fontsize=30)

plt.show()

plt.imshow(imgr,cmap=plt.cm.gray)
plt.title('Reconstructed Image',fontsize=30)

```
