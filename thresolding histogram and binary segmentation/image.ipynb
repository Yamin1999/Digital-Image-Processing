# thresolding and binary segmentation

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img = plt.imread("ruet2.jpg")
```

```python
plt.imshow(img)
```

```python
imb3 = img[:,:,0]
```

```python
imb3.shape
```

```python
plt.imshow(imb3, cmap='gray')
```

```python
imb3 = np.int16(np.fix(imb3))
```

```python
imb3.shape
```

```python
m,n =imb3.shape
```

```python
hist = np.zeros(256)
```

```python
for i in range(m):
    for j in range(n):
        k = imb3[i][j]
        hist[k]=hist[k]+1
```

```python
hist
```

```python
plt.plot(hist)
```

```python
for i in range(imb3.shape[0]):
    for j in range(imb3.shape[1]):
        if (imb3[i][j]>180):
            imb3[i][j]=1
        else:
            imb3[i][j]=0
```
